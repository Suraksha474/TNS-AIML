**Dictionaries**

```python
# 1. Program that takes a sentence and creates a dictionary with word counts

sentence = input("Enter a sentence: ")
words = sentence.split()
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
print("Word count dictionary:", word_count)
```

```python
# 2. Program to map roll numbers to names using two lists and form a dictionary

roll_numbers = input("Enter roll numbers separated by space: ").split()
names = input("Enter names separated by space: ").split()
mapping = dict(zip(roll_numbers, names))
print("Roll number to name mapping:", mapping)
```

```python
# 3. Program that takes a dictionary of students and their marks, and prints the name(s) of the student(s) with the highest marks

students = {"Alice": 85, "Bob": 98, "Charlie": 90, "David": 70}
max_marks = max(students.values())
top_students = [name for name, marks in students.items() if marks == max_marks]
print("Student(s) with highest marks:", top_students)
```
