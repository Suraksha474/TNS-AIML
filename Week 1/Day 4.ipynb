**List**

```python
# 1. Program to find the difference between the maximum and minimum elements in a list

numbers = [1, 15, 25, 12, 10]
diff = max(numbers) - min(numbers)
print("Difference between max and min:", diff)
```

```python
# 2. Program to remove all elements at odd indices from a list

lst = [1, 9, 3, 6, 5, 8, 7, 12]
result = [lst[i] for i in range(len(lst)) if i % 2 == 0]
print("List after removing elements at odd indices:", result)
```

```python
# 3. Program that takes a list of integers and prints only the elements that appear exactly once

nums = [6, 2, 2, 3, 4, 5, 5, 6, 7, 8]
unique = [x for x in nums if nums.count(x) == 1]
print("Elements that appear exactly once:", unique)
```
