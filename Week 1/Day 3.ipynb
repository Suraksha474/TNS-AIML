**Functions and String Handling**

```python
# 1. Function to check whether a given string is a pangram (contains all letters of the alphabet)
import string

def is_pangram(s):
    alphabet = set(string.ascii_lowercase)
    return set(s.lower()) >= alphabet

# Example usage:
text = input("Enter a string to check for pangram: ")
if is_pangram(text):
    print("The string is a pangram.")
else:
    print("The string is not a pangram.")
```

```python
# 2. Function that takes a sentence and returns the word with the maximum length

def max_length_word(sentence):
    words = sentence.split()
    return max(words, key=len) if words else ""

# Example usage:
sentence = input("Enter a sentence: ")
print("Longest word:", max_length_word(sentence))
```

```python
# 3. Function to count the number of uppercase and lowercase characters in a string

def count_case(s):
    upper = sum(1 for c in s if c.isupper())
    lower = sum(1 for c in s if c.islower())
    return upper, lower

# Example usage:
s = input("Enter a string to count cases: ")
upper, lower = count_case(s)
print(f"Uppercase: {upper}, Lowercase: {lower}")
```
