**Tuple**

```python
# 1. Take 5 numbers from the user, store in a tuple, print max and min

nums = tuple(int(input(f"Enter number {i+1}: ")) for i in range(5))
print("Tuple:", nums)
print("Maximum:", max(nums))
print("Minimum:", min(nums))
```

```python
# 2. Count how many elements in a tuple are divisible by 3

tup = (3, 6, 9, 10, 12, 11, 15,20)
count = sum(1 for x in tup if x % 3 == 0)
print("Count of elements divisible by 3:", count)
```

```python
# 3. Reverse the contents of a tuple without using reversed()

tup = (10, 2, 9, 7, 15, 4, 1)
reversed_tup = tup[::-1]
print("Reversed tuple:", reversed_tup)
```
