**Set**

```python
# 1. Take two lists from the user and print the common elements using sets

list1 = input("Enter elements of first list separated by space: ").split()
list2 = input("Enter elements of second list separated by space: ").split()
common = set(list1) & set(list2)
print("Common elements:", list(common))
```

```python
# 2. Check if two sets are disjoint or not

set1 = set(input("Enter elements of first set separated by space: ").split())
set2 = set(input("Enter elements of second set separated by space: ").split())
if set1.isdisjoint(set2):
    print("The sets are disjoint.")
else:
    print("The sets are not disjoint.")
```

```python
# 3. Find all unique vowels present in a given string using set

s = input("Enter a string: ")
vowels = set("aeiouAEIOU")
found = set(c for c in s if c in vowels)
print("Unique vowels present:", found)
```
