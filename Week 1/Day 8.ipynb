```python
# 1. Convert a list of tuples into a dictionary

tuple_list = [('a', 40), ('b', 20), ('c', 70)]
result_dict = dict(tuple_list)
print("Dictionary:", result_dict)
```

```python
# 2. Flatten all values in a dictionary with lists as values into a single list

d = {'x': [10, 22], 'y': [35, 40], 'z': [7]}
flat_list = [item for sublist in d.values() for item in sublist]
print("Flattened list:", flat_list)
```

```python
# 3. Return a dictionary with elements as keys and their frequency as values from a list with duplicates

lst = [1, 2, 2, 3, 3, 3, 4, 5, 5, 6]
freq_dict = {}
for item in lst:
    freq_dict[item] = freq_dict.get(item, 0) + 1
print("Frequency dictionary:", freq_dict)
```
